## Task 5 :- Traffic sign classification

#### Step 1:- Importing all the required modules.................

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image 
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D , Flatten, BatchNormalization
print(tf.__version__)
```

#### Step 2:- Prepare the German Traffic Sign Dataset For Training...............

```python
train_dir = os.path.join('../input/gtsrb-german-traffic-sign/Train/')
validation_dir = os.path.join('./Data/test/')

train_datagen = ImageDataGenerator(rescale = 1./255,)


train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size = (64,64),
                                                    batch_size = 250,
                                                    class_mode = 'categorical')
```

#### Step 3:- Build the Deep Learning Model................

```python
model = Sequential([
    Conv2D(16,(5,5),padding = 'same',activation='relu',input_shape = (64,64,3)),
    Conv2D(16,(5,5),padding = 'same',activation='relu'),
    MaxPool2D((2,2)),
    Conv2D(16,(3,3),padding = 'same',activation = 'relu'),
    Conv2D(16,(3,3),padding = 'same',activation = 'relu'),
    Conv2D(16,(3,3),padding = 'same',activation = 'relu'),
    MaxPool2D((2,2)),
    Conv2D(32,(3,3),padding = 'same',activation = 'relu'),
    Conv2D(32,(3,3),padding = 'same',activation = 'relu'),
    Conv2D(32,(3,3),padding = 'same',activation = 'relu'),
    MaxPool2D((2,2)),
    Conv2D(64,(3,3),padding = 'same',activation = 'relu'),
    Conv2D(64,(3,3),padding = 'same',activation = 'relu'),
    Conv2D(64,(3,3),padding = 'same',activation = 'relu'),
    MaxPool2D((2,2)),
    Conv2D(128,(3,3),padding = 'same',activation = 'relu'),
    Conv2D(128,(3,3),padding = 'same',activation = 'relu'),
    Conv2D(128,(3,3),padding = 'same',activation = 'relu'),
    MaxPool2D((2,2)),
    Flatten(),
    Dense(256,activation = 'relu'),
    Dense(43,activation = 'softmax')
])
```

#### Step 4:- Visualize the Model.................

```python
model.summary()
```

#### Step 5:- Model Compilation..............

```python
model.compile(loss = 'categorical_crossentropy',
              optimizer = 'adam',
             metrics = ['acc'])
```

#### Step 6:- Training......................... 

```python
history = model.fit_generator(train_generator,
                             epochs = 20,
                             verbose = 2)
```

#### Step 8:- Training Loss and Accuracy Visualization..............

```python
import matplotlib.pyplot as plt
acc = history.history['acc']
loss = history.history['loss']
epochs = range(len(acc))
plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, loss, 'b', label='Training loss')
plt.title('Training accuracy and loss')
plt.legend(loc=0)
plt.figure()
plt.show()
```
